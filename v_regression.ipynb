```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Virat-Kohli-International-Cricket-Centuries.csv") 
```

```python
df.columns.str.strip()
```

```python
df = df.drop(['Date' , 'Venue'], axis=1)
```

<!-- Regression  -->

```python
df['Result'] = df['Result'].str.strip()  # Remove leading/trailing spaces
# df["Result"] = (df["Result"] == "Won").astype(int)  # Convert to 0 and 1
```

```python
for label in df.columns[2:]:
    if label == "Result":
        # Define colors for each result category
        colors = {"Won": "green", "Lost": "red", "Tied": "blue", "Drawn": "orange"}
        
        for result, color in colors.items():
            subset = df[df["Result"] == result]
            plt.scatter(subset[label], subset["Runs"], label=result, color=color)
        
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=4)
    else:
        plt.scatter(df[label], df["Runs"])
    
    plt.title(label)
    plt.ylabel("Runs")
    plt.xlabel(label)
    
    if label == "Against":
        plt.xticks(rotation=45)
    elif label == "Venue":
        plt.xticks(rotation=90, fontsize=6)
    elif label == "Date":
        plt.xticks(rotation=90, fontsize=4)
        continue
    plt.show()
```
